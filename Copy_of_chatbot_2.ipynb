```python
loader=CSVLoader("/content/drive/MyDrive/codebasics/codebasics1.csv",encoding="cp1252",source_column="prompt")
data=loader.load()
data[0]
```

```python
!pip install InstructorEmbedding
```

```python
ib = HuggingFaceInstructEmbeddings()
```

```python
pip install faiss-gpu
```

```python
vectordb=FAISS.from_documents(data,ib)
vectordb
```

```python
retriver=vectordb.as_retriever(score_threshold=0.8)
retriver.get_relevant_documents("I have never done programming in my life. Can I take this bootcamp?")
```

```python
prompt_template="""Given the following context and a question,generate an based on this context only.
In the answer try to provide as much text as possible from making much changes.If the answer is not found in the context,kindly
state 'i dont know'.Dont try to make up an answer.
CONTEXT:{context}
QUESTION:{question}"""
```

```python
prompt=PromptTemplate(template=prompt_template,input_variables=["content","question"])
```

```python
chain=RetrievalQA.from_chain_type(llm=llm,
                                  chain_type="stuff",
                                  retriever=retriver,  # Pass the retriever object, not a string
                                  input_key="query",
                                  return_source_documents=True,
                                  chain_type_kwargs={"prompt":prompt})
```

```python
a=chain("what is the placement guarantee")
b=a["result"]
b
```

```python
from transformers import pipeline

pipe = pipeline("text-to-speech", model="suno/bark-small")
text = b
output = pipe(text)
```

```python
from IPython.display import Audio
Audio(output["audio"], rate=output["sampling_rate"])
```
